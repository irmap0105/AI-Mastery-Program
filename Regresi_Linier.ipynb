```python
data = pd.read_csv("house_prices.csv")
data.head()
```

```python
plt.scatter(data['LotArea'],data['SalePrice'])
plt.xlabel('Area')
plt.ylabel('Price')
```

```python
X=data.LotArea[:,np.newaxis]
y=data.SalePrice
```

```python
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
y_pred = regressor.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X, y_pred,"r-")
plt.xlabel('Area')
plt.ylabel('Price')
```
